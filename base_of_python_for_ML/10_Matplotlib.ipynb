# Pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2,10)
y=x**2
print(x)
```

```python
plt.plot(x,y, c="black",lw=2,ls="--") #y en fonction de x de couleur noir, épaisseur 2 en trétillait (pas nécessaire)
plt.show() #afficher le graphique
#même dimension nécessaire!!
```

```python
plt.scatter(x,y) #nuage de points
```

# Marche à suivre

```python
plt.figure(figsize=(12,8)) #fenêtre de travail de dimension 12,8 inches
plt.plot(x,y, label="quadratique") #graphique avec le nom quadratic
plt.plot(x,x**3, label = "cubic")
plt.title("figure 1") #titre de la figure
plt.xlabel("axe x") #titre axe x
plt.ylabel("axe y") #titre axe y
plt.legend()
plt.show()
plt.savefig("figure.png")
```

```python
plt.figure()
plt.subplot(2,1,1) #lignes, colonnes, position
plt.plot(x,y, c="red")
plt.subplot(2,1,2) #afficher plusieurs graphiques sur la même fenêtre
plt.plot(x,x**3)
plt.show()
```

```python
#POO (moins utile)
fig,ax = plt.subplots(2,1,sharex=True) #avoir plusieurs graphique
ax[0].plot(x,y)
ax[1].plot(x, np.sin(x))
plt.show()
```

## Exercice

```python
dataset = {f"experience{i}":np.random.randn(100,3) for i in range(4)} #3 variables avec 100 données
```

```python
def graphique(data):
    n=len(data)
    plt.figure(figsize=(12,8))
    for k,i in zip(data.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
    
```

```python
graphique(dataset)
```

# Graphiques


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Classification

```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target
names = list(iris.target_names)
```

```python
plt.scatter(x[:, 0],x[:,1], c=y, alpha=0.5, s=x[:,2]*100) #pour 2 variables (alpha = transparence, s = taille des pointss
plt.xlabel("longueur sépal")
plt.ylabel("largeur sépal")
```

## Graphique 3D


```python

#visualiser le graphe (zoomer, dezoomer)
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection="3d") #création des axes
ax.scatter(x[:,0], x[:,1], x[:,2],c = y) # 3 variables
```

```python
f = lambda x, y: np.sin(x)+np.cos(x+y)
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y) #créer une grille à partir des 2 axes
Z = f(X,Y)
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z, cmap="summer") #cmap=couleur de la surface
#https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
```

## Histogramme

```python
plt.hist(x[:,0], bins=20) #bins=nbr de section par histogram
plt.hist(x[:,1], bins=20)
```

```python
plt.hist2d(x[:,0],x[:,1], cmap="Blues")
plt.colorbar() #correspondance des couleurs
```

```python
from scipy import misc
face = misc.face(gray=True)
#plt.imshow(face, cmap="gray")
plt.hist(face.ravel(), bins = 255) #fréquence des couleurs des pixels
plt.show()
```

## Contour plots

```python
plt.contour(X,Y,Z,20, colors="blue") #vue du dessus avec 20 niveau
```

```python
plt.contourf(X,Y,Z,20, cmap="Blues")
plt.colorbar()
```
