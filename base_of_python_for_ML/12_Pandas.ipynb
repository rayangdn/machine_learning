### Resample

```python
#regrouper nos données selon une fréquence
bitcoin.loc["2019", "Close"].resample("M").plot() #par mois "W" par week, ...
plt.show()
```

```python
bitcoin.loc["2019", "Close"].resample("M").mean().plot() #moyenne sur chaque mois
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc["2019", "Close"].plot()
bitcoin.loc["2019", "Close"].resample("M").mean().plot(label="moyenne par mois", ls=":",lw=3,alpha=0.8)
bitcoin.loc["2019", "Close"].resample("W").mean().plot(label="moyenne par semaine", ls="--",lw=3,alpha=0.8)
plt.legend()
plt.show()
```

```python
m = bitcoin.loc["2019", "Close"].resample("W").agg(["mean","std","max", "min"])
```

```python
plt.figure(figsize=(12,8))
m["mean"]["2019"].plot(label="moyenne par semaine")
plt.fill_between(m.index, m["max"], m["min"], alpha=0.2, label="min-max par semaine")
#zone entre max et min
plt.legend()
plt.show()
```

### Moving Average
moyenne sur une fenêtre de valeur

```python
plt.figure(figsize=(12,8))
bitcoin.loc["2019-09", "Close"].plot()
bitcoin.loc["2019-09", "Close"].rolling(window=7).mean().plot(label="moving average", lw=3, ls=":")
#fenêtre de 7 jours
bitcoin.loc["2019-09", "Close"].rolling(window=7, center=True).mean().plot(label="moving average", lw=3, ls=":")
#fenêtre centrée
plt.legend()
plt.show()
```

### Exp Weighted Function
moyenne mobile exponentielle

```python
plt.figure(figsize=(12,8))
bitcoin.loc["2019-09", "Close"].plot()
for i in np.arange(0.2,1,0.2): #comparer les valeurs de alpha
    bitcoin.loc["2019-09", "Close"].ewm(alpha=i).mean().plot(label=f"EWM {i}", lw=1, ls="--")
plt.legend()
plt.show()
```

# Assembler des datasets

```python
ethereum = pd.read_csv("ETH-EUR.csv", index_col="Date", parse_dates=True)
```

```python
pd.merge(bitcoin, ethereum, on="Date", how="outer", suffixes=("_btc","_eth")) #date est la m^me pour les 2, on assemble selon celle-çi, how est la manière de l'assembler
#outer: commence à la plus ancienne
btc_eth = pd.merge(bitcoin, ethereum, on="Date", how="inner", suffixes=("_btc","_eth"))
#inner : commence à la plus récente
```

# Exercice : Trading Strategy

### Turtle strategy

```python
data = bitcoin.copy() #copier les datas du bitcoin
data["Buy"]=np.zeros(len(bitcoin))
data["Sell"]=np.zeros(len(bitcoin))
```

```python
data["RollingMax"]= data["Close"].shift(1).rolling(window=28).max() #shift pour décaller de 1 jour
data["RollingMin"]= data["Close"].shift(1).rolling(window=28).min()
data.loc[data["RollingMax"] < data["Close"], "Buy"] = 1
data.loc[data["RollingMin"] > data["Close"], "Sell"] = -1
```

```python
start = "2019"
end = "2019"
fig, ax = plt.subplots(2, figsize=(12,8), sharex= True)
ax[0].plot(data["Close"][start:end])
ax[0].plot(data["RollingMin"][start:end])
ax[0].plot(data["RollingMax"][start:end])
ax[0].legend(["close", "min", "max"])
ax[1].plot(data["Buy"][start:end], c="g")
ax[1].plot(data["Sell"][start:end], c="r")
ax[1].legend(["buy", "sell"])
```
