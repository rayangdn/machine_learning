# Exercice

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
face.shape
```

```python
#kind of compression d'image
face = face[::2,::2]
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
h = face.shape[0]
w = face.shape[1]
zoom_face = face[h//4:-h//4,w//4:-w//4]
zoom_face[zoom_face>150]=255
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

## NAN Corrections

```python
A = np.random.randn(5,5)
A[0,2] = np.nan #not a number (peut arriver dans un datasets)
A[4,3]= np.nan
A
```

```python
print(np.nanmean(A)) #moyenne sans inclure les nan values
print(np.isnan(A)) #donne si il y'a des nan
np.isnan(A).sum() #sum des nan
```

```python
A[np.isnan(A)] = 0 #si il y'a des nan on remplace par zéros
A
```
