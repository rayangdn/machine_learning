# 1. Exploratory Data Analysis

## General info

```python
train_data_cl = train_data.copy()
test_data_cl = test_data.copy()
y_train = train_data_cl["SalePrice"]
#Dropping columns to concat them
train_data_cl.drop(["Id","SalePrice"], axis=1, inplace=True)
test_data_cl.drop("Id",axis=1, inplace=True)
```

```python
df = pd.concat([train_data_cl,test_data_cl], axis=0, ignore_index=True)
```

```python
print(df.shape) #we have full datas to analyse
```

```python
df.dtypes.value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.isna(), cbar=False) #To see where we have the most Na values
```

```python
missing_values = df.isna().sum()/df.shape[0] #To see the % of Na
print(missing_values[missing_values > 0].sort_values(ascending=False))
#Need to be carefull because most of the Na values correspond to "None" and is not actually a missing value
```

## Target Column

```python
plt.figure(figsize=(20,10))
sns.histplot(x=y_train)
plt.show()
print("mean :", y_train.mean())
print("std :", y_train.std())
print("median :", y_train.median())
print("mode :", y_train.mode()[0])
```

## Categorical Variables

```python
for col in df.select_dtypes("object"): #choisir un type 
    print(f"{col :-<40}{df[col].unique()}") #astuce pour rendre jolie
        
```

```python
for col in df.select_dtypes("object"): #choisir un type 
    plt.figure()
    df[col].value_counts().plot.pie()
```

## Continuous Variables

```python
for col in df.select_dtypes(["float","int"]):
    plt.figure()
    sns.kdeplot(df[col], shade=True)
```
